```python
import numpy as np 
import pandas as pd 
```

```python
hairLoss = pd.read_csv("Predict Hair Fall.csv")
```

```python
hairLoss.head()
```

```python
hairLoss['EnvironmentalFactors'].value_counts()
```

```python
hairLoss['MedicalConditions'].value_counts()
```

```python
hairLoss = hairLoss.drop(columns="Id",axis=1)
```

```python
hairLoss.head()
```

```python
target_mean = hairLoss.groupby('MedicationsandTreatments')['HairLoss'].mean()
```

```python
target_mean
```

```python
hairLoss['MedicationsandTreatments'] = hairLoss['MedicationsandTreatments'].map(target_mean)
```

```python
hairLoss.head()
```

```python
target_mean = hairLoss.groupby('NutritionalDeficiencies')['HairLoss'].mean()
```

```python
target_mean
```

```python
hairLoss['NutritionalDeficiencies'] = hairLoss['NutritionalDeficiencies'].map(target_mean)
```

```python
hairLoss.head()
```

```python
target_mean = hairLoss.groupby('MedicalConditions')['HairLoss'].mean()
```

```python
target_mean
```

```python
hairLoss['MedicalConditions'] = hairLoss['MedicalConditions'].map(target_mean)
```

```python
hairLoss.head()
```

```python
target_mean = hairLoss.groupby('Stress')['HairLoss'].mean()
```

```python
target_mean
```

```python
hairLoss['Stress'] = hairLoss['Stress'].map(target_mean)
```

```python
hairLoss.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label = LabelEncoder()
```

```python
hairLoss["Genetics"] = label.fit_transform(hairLoss["Genetics"])
hairLoss["HormonalChanges"] = label.fit_transform(hairLoss["HormonalChanges"])
hairLoss["PoorHairCareHabits"] = label.fit_transform(hairLoss["PoorHairCareHabits"])
hairLoss["EnvironmentalFactors"] = label.fit_transform(hairLoss["EnvironmentalFactors"])
hairLoss["Smoking"] = label.fit_transform(hairLoss["Smoking"])
hairLoss["WeightLoss"] = label.fit_transform(hairLoss["WeightLoss"])
```

```python
hairLoss.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
```

```python
# standardising the data 
x = hairLoss.drop(columns="HairLoss", axis=1)
y = hairLoss.iloc[:,10]
```

```python
x
```

```python
scale.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test , y_train , y_test = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
print(x.shape)
print(y.shape)
```

```python
y_pre = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pre)
```

```python
import pickle 
```

```python
filename = "trained_modelNew.sav"
pickle.dump(model,open(filename,"wb"))
```

```python
loaded_model = pickle.load(open("trained_modelNew.sav","rb"))
```

```python
input_data = (0,0,0.500000,0.490196,0.512821,0.518519,30,0,1,1,0)

# changing the input data to numpy array

input_data_numpy_array = np.asarray(input_data)

# reshape the array as we predicted for one instance

input_data_reshaped = input_data_numpy_array.reshape(1,-1)

prediction = loaded_model.predict(input_data_reshaped)
print(prediction)

if(prediction[0] == 0):
    print("Has no Hair Loss")

else:
    print("Has Hair Loss")
```
